```python
import numpy as np

MatA = np.array([[2, 4, 6],
                 [8, 10, 12]])

MatB = np.array([[3, 3, 3],
                 [6, 6, 6],
                 [9, 9, 9]])

print("Matrix A:\n", MatA)
print("Matrix B:\n", MatB)

print("Shape of A:", MatA.shape)
print("Shape of B:", MatB.shape)

productAB = np.dot(MatA, MatB)
print("Product of A and B:\n", productAB)
```

```python
import numpy as np

mat_a = np.array([[1, 2], [3, 4]])
mat_b = np.array([[5, 6], [7, 8]])

result_add = np.add(mat_a, mat_b)
result_sub = np.subtract(mat_a, mat_b)
res_elem = np.multiply(mat_a, mat_b)
res_dot = np.dot(mat_a, mat_b)
res_trans = np.transpose(mat_a)

print("Sum:", result_add)
print("Difference:", result_sub)
print("Product:", res_elem)
print("Dot:", res_dot)
print("Transpose of Matrix:", res_trans)
```
